```python
import numpy as np

# Load the original image
original_image = Image.open('./PicsArt_09-08-02.20.25 (2).png')

# Enhance the image resolution
enhancement_factor = 2  # Adjust this value as needed
enhanced_image = original_image.resize(
    (original_image.width * enhancement_factor, original_image.height * enhancement_factor),
    Image.BICUBIC  # You can also try other resampling methods
)

# Convert both images to grayscale (L mode) for comparison
original_gray = original_image.convert('L')
enhanced_gray = enhanced_image.convert('L')

# Convert to NumPy arrays
original_array = np.array(original_gray)
enhanced_array = np.array(enhanced_gray)

# Calculate the Mean Squared Error (MSE)
mse = np.mean((original_array - enhanced_array) ** 2)

# Calculate the percentage improvement
percentage_improvement = ((np.mean(original_array ** 2) - mse) / np.mean(original_array ** 2)) * 100

print(f"Mean Squared Error (MSE): {mse}")
print(f"Percentage Improvement: {percentage_improvement:.2f}%")
```

```python
import numpy as np

# Load the original image
original_image = Image.open('./PicsArt_09-08-02.20.25 (2).png')

# Enhance the image resolution
enhancement_factor = 2  # Adjust this value as needed
enhanced_image = original_image.resize(
    (original_image.width * enhancement_factor, original_image.height * enhancement_factor),
    Image.BICUBIC  # You can also try other resampling methods
)

# Resize the original image to the same dimensions as the enhanced image
original_image_resized = original_image.resize(enhanced_image.size, Image.BICUBIC)

# Convert both images to grayscale (L mode) for comparison
original_gray = original_image_resized.convert('L')
enhanced_gray = enhanced_image.convert('L')

# Convert to NumPy arrays
original_array = np.array(original_gray)
enhanced_array = np.array(enhanced_gray)

# Calculate the Mean Squared Error (MSE)
mse = np.mean((original_array - enhanced_array) ** 2)

# Calculate the percentage improvement
percentage_improvement = ((np.mean(original_array ** 2) - mse) / np.mean(original_array ** 2)) * 100

print(f"Mean Squared Error (MSE): {mse}")
print(f"Percentage Improvement: {percentage_improvement:.2f}%")
```

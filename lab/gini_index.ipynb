```python
# Importing necessary libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
import matplotlib.pyplot as plt
```

```python
# Load the dataset
file_path = './play_tennis_train.csv'  # Replace this with your file path if needed
data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
print("Dataset Preview:")
display(data.head())
```

```python
# Dropping 'day' as it is not a feature
data = data.drop(columns=["day"])
```

```python
# Encoding categorical variables
data_encoded = pd.get_dummies(data, columns=["outlook", "temp", "humidity", "wind"], drop_first=True)
X = data_encoded.drop(columns=["play"])
y = data_encoded["play"].map({"Yes": 1, "No": 0})  # Encoding target variable
```

```python
# Train a Decision Tree Classifier using Gini Index
clf = DecisionTreeClassifier(criterion="gini", random_state=0)
clf.fit(X, y)
```

```python
# Extract and display the decision tree rules
tree_rules = export_text(clf, feature_names=list(X.columns))
print("Decision Tree Rules:")
print(tree_rules)
```
